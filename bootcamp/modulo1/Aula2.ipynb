## Caderno de exploração dos dados do datasus 2008 pra frente

```python
import pandas as pd
```

```python
dados = pd.read_csv('A182305189_28_143_208.csv', #endereço do dataframe
                    thousands='.', #separador de milhar por ponto
                    decimal=',', #separador de decimal é por virgula
                    encoding='ISO-8859-1', #o codec do arquivo é iso-8859-1
                    skipfooter=12, # iginora as ultimas 12 linhas do arquivo
                    skiprows=3, # ignora as 3 primeiras linhas do arquivo
                    sep=';', # separador é por ponto e virgula
                    engine='python' ) 
dados.head()
```

```python
dados.mean()
```

```python
dados.info()
```

```python
pd.options.display.float_format = "{:.2f}".format #configura a impressão global para 2 casas decimais
```

```python
dados.plot(x='Unidade da Federação', 
           y='2008/Ago', 
           kind='bar', 
           figsize=(10,7))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as tiker

ax = dados.plot(x='Unidade da Federação', 
                y='2008/Ago', 
                kind='bar', 
                figsize=(10,7))
ax.yaxis.set_major_formatter(tiker.StrMethodFormatter('{x:,.0f}') )
#https://matplotlib.org/3.2.1/gallery/ticks_and_spines/tick-formatters.html

plt.title('Valor por unidade da federação')
plt.show()
```

```python
dados_ago_08 = dados.sort_values(by=['2008/Ago'], ascending=True)
```

```python
ax = dados_ago_08.plot(x='Unidade da Federação', 
                       y='2008/Ago', 
                       kind='barh', 
                       figsize=(10,7))
ax.xaxis.set_major_formatter(tiker.StrMethodFormatter('{x:,.0f}') )
#https://matplotlib.org/3.2.1/gallery/ticks_and_spines/tick-formatters.html

plt.title('Valor por unidade da federação')
plt.show()
```

## Desafio: faça a mesma análise pro mês mais recente.

```python
dados_dez_16 = dados.sort_values(by=['2016/Dez'], ascending=True)
ax = dados_dez_16.plot(x='Unidade da Federação', 
                       y='2016/Dez' ,
                       kind='barh', 
                       figsize=(15,7), width=0.8)

ax.xaxis.set_major_formatter(tiker.StrMethodFormatter('{x:,.0f}') )
#https://matplotlib.org/3.2.1/gallery/ticks_and_spines/tick-formatters.html

plt.title('Valor gasto por estado em Dezembro de 2016', fontsize=16)
plt.show()
```

```python
#pegar dados só de onde tem dados preenchidos#
##############################################

dados.mean() # a média só retorna resultado de onde encontrou dados preenchidos
```

```python
colunas_usaveis = dados.mean().index.tolist() #retorna somente as colunas que tem dados preenchidos
colunas_usaveis.insert(0,'Unidade da Federação')
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```python
usaveis = usaveis.set_index('Unidade da Federação')
usaveis.head()
```

```python
usaveis['2019/Ago'].head()
```

```python
usaveis.loc['12 Acre']
```

```python
usaveis.T.head()
```

```python
usaveis = usaveis.drop('Total', axis=1)
```

```python
ax = usaveis.T.plot(figsize=(10,6))
plt.show(ax)
```

## Desafio: reposicionar a legenda. Dentro? Fora? Onde?

```python
ax = usaveis.T.plot(figsize=(10,7))
ax.legend(loc='upper center', bbox_to_anchor=(1.2, 1), shadow=True, ncol=1)
ax.yaxis.set_major_formatter(tiker.StrMethodFormatter('{x:,.0f}') )
plt.title('Gasto dos Estados por Mês', fontsize=16)
plt.show(ax)
```
